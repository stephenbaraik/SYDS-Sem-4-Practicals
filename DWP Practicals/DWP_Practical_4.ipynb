```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler,StandardScaler,MaxAbsScaler,RobustScaler,Normalizer
```

**Normalization is scaling technique in which values are shifted and rescaled ranging between 0 & 1 aka min/Max Scalar**

x' = x - Xmin / Xmax - Xmin

Standardisation is the technique 
when the values are centered around the mean with a unique standard deviation 

X' = x - mu/sigma


if our data is not normally distributed than we use normalization else we use normal distribution



```python
df = pd.read_csv(r"https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv")
df
```

```python
df.describe()
```

```python
scalar = MinMaxScaler()
# normalize the data 
normalized_data = scalar.fit_transform(df)
```

```python
df2 = pd.DataFrame(normalized_data,columns=df.columns)
df2.describe()
```

```python
scalar = StandardScaler()
df_scaled = scalar.fit_transform(df)
df3 = pd.DataFrame(df_scaled,columns=df.columns)
df3
```

```python
# mean Normalization
df4 = (df - df.mean())/df.max() - df.min()
df4
```

```python
# maximum absolute scaling
df5 = MaxAbsScaler().fit_transform(df)
df5 = pd.DataFrame(df5,columns=df.columns)
df5
```

```python
# median and quantile normalization scaling
# x - median(x) / iqr(x)
df6 = RobustScaler().fit_transform(df)
df6 = pd.DataFrame(df6,columns=df.columns)
df6
```

```python
# L2 regulization is also called as ridge regression it prevents overfitting problem by adding a penalty to the Loss function,
# this penalty discourages large cooeficients in the model.

# vector unit length scaling using L2 normalization 
df7 = Normalizer(norm = 'l2').fit_transform(df)
df7 = pd.DataFrame(df7,columns=df.columns)
df7
```
